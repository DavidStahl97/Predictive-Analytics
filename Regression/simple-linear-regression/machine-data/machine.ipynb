```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('machinedata.csv', index_col='index')
data
```

```python
data.corr()
```

```python
y = data['weight'].values
X_length = data['length'].values
X_height = data['height'].values
```

```python
from sklearn.linear_model import LinearRegression

regressor_length = LinearRegression()
regressor_length.fit(X_length.reshape(-1, 1), y)

regressor_height = LinearRegression()
regressor_height.fit(X_height.reshape(-1, 1), y)
```

```python
plt.scatter(X_length, y)
plt.plot(X_length, regressor_length.predict(X_length.reshape(-1, 1)), color = 'blue')
plt.title('weight vs length')
plt.xlabel('length')
plt.ylabel('weight')
```

```python
plt.scatter(X_height, y)
plt.plot(X_height, regressor_height.predict(X_height.reshape(-1, 1)), color = 'blue')
plt.title('weight vs height')
plt.xlabel('length')
plt.ylabel('height')
```
